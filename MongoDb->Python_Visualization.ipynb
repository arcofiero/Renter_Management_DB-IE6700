```python

import pandas as pd
import plotly.graph_objects as go

fig = go.Figure(go.Scatter(x = df_renter['F_Name'], y = df_renter['Rent_Amount'],
                  name='House rent (IN USD)'))

fig.update_layout(title='Rent Amount per Renters',
                   plot_bgcolor='rgb(230, 230,230)',
                   showlegend=True)

fig.show()
```

```python
import plotly.express as px
fig1 = px.pie(df_renter, values="R_ID",
             names="R_Age", title ="Renter Age Distirbution Pie",
             color_discrete_sequence=px.colors.sequential.RdBu)


fig1.show()
```

```python
import plotly.express as px

fig = px.bar_polar(df_renter, r="Rent_Amount", theta="F_Name", color="R_Age",
                   color_discrete_sequence= px.colors.sequential.Plasma_r,
                   title="Proportion of Rent Amount per renter vs Age"
                  )
fig.show()
```
